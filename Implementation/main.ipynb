```python
import json
from QRSDetection import*
from QRSCalculator import*

import pandas as pd
import numpy as np
import joblib
```

```python
# -------------------------
#          LOAD DATA 
# -------------------------
df = pd.read_excel("Data/Elderly_Query.xlsx")
df.head(3)
```

```python
# initialize calculator
qrs_model = QRSCalculator()

# store results
results = []
# -------------------------
# MAIN LOOP
# -------------------------
for query in df["Query"]:

    # Step 1: detect scores using LLM
    scores = detect_qrs_scores(query)

    # Step 2: calculate QRS
    qrs_result = qrs_model.evaluate(scores)

    # combine everything
    row = {
        "Query": query,
        "MS": scores["MS"],
        "SC": scores["SC"],
        "MD": scores["MD"],
        "UE": scores["UE"],
        "PA": scores["PA"],
        "ES": scores["ES"],
        "QRS": qrs_result["QRS"],
        "Risk_Level": qrs_result["Risk_Level"]
    }

    results.append(row)

# -------------------------
# SAVE RESULTS
# -------------------------
results_df = pd.DataFrame(results)

results_df.to_csv("Results/QRS_results.csv", index=False)

print("Results saved to QRS_results.csv")

print(results_df.head())
```
